```python
# Non-delivered orders are the orders that have order_purchase_timestamp but have no order_delivered_customer_date

# Let's analyze all created orders and find out which are not delivered to a customer
# 1. choose only non-delivered orders - orders without order_delivered_customer_date
# 2. group by year and month of order_purchase_timestamp
# 3. count numbers of non-delivered orders each month
# 4. find mean value of non-delivered orders per month

b = orders_df \
    .query("order_delivered_customer_date.isna()") \
    .groupby(['purchase_year', 
              'purchase_month'])['order_id'] \
    .count().mean().round(0) 
print('По разным причинам каждый месяц не доставляется в среднем', b, 'заказа.')    
```

```python
# Задание 5. 
# Используя pandas, проведи когортный анализ пользователей. 
# В период с января по декабрь выяви когорту с самым высоким retention на 3й месяц. 
```

```python
# Когорта — это группа людей, которая совершила нужное действие (action) в определенный промежуток времени (period).
# Когортный анализ — это наблюдение за когортами по одной или нескольким метрикам (KPI) 
# на временных отрезках заданной длительности (time). 
```

```python
# Retention rate (RR) показывает отношение клиентов, которые вернулись за повторными покупками, 
# к числу клиентов на начало расчетного периода.
```

```python
# В нашем задании:
# Action = клиент создает заказ
# Period = месяц
# Time = с января по декабрь
# KPI = RR - доля повторных клиентов
```

```python
# для решения задачи будем использовать объединенную таблицу клиентов и заказов из предыдущего задания
d.head()
# в этом задании будем анализировать клиентов, создавших заказ, без условия его оплаты
```

```python
# определим месяц и год создания первого заказа по каждому клиенту
d['order_period'] = d.order_purchase_timestamp.apply(lambda x:x.strftime('%Y-%m'))
```

```python
d.head()
```

```python
# определим месяц и год первого заказа по каждому клиенту 
first_order = d \
    .groupby('customer_unique_id', as_index=False) \
    .agg({'order_period':'min'}) \
    .rename(columns = {'order_period':'first_order'})
first_order
```

```python
# добавим дату первой покупки в исходный датафрейм
orders_merged = d.merge(first_order, on = 'customer_unique_id')
```

```python
# сгруппируем уникальных пользователей сначала по месяцу первой покупки, а потом по месяцу следующей покупки
# при этом ограничим дату первой покупки январем 2017 года, а дату повторных заказов - декабрем 2017, 
# так как это заданный в условии период для анализа.
cohorts = orders_merged \
    .query("first_order >= '2017-01' and order_period <= '2017-12'") \
    .groupby(['first_order', 'order_period'], as_index=True) \
    .agg({'customer_unique_id':'nunique'})
cohorts
# в нашем случае когортами являются строки колонки 'first_order'
```

```python
# создадим функцию для нумерации отрезков времени (месяцев) анализа когорт
def cohort_period(df):
    df['cohort_period'] = np.arange(len(df)) + 1
    return df
```

```python
# создадим колонку с порядковым номером месяца анализа метрики когорты
cohorts = cohorts.groupby('first_order').apply(cohort_period)
cohorts
```

```python
# создадим таблицу с размером каждой когорты (количество клиентов, сделавших первый заказ)
cohorts_size = cohorts.groupby('first_order')['customer_unique_id'].first()
cohorts_size
```

```python
# представим данные по когортам в виде сводной таблицы, предварительно сбросив индекс
cohorts_pivot = cohorts.reset_index().pivot(  index = 'first_order', 
                                            columns = 'cohort_period', 
                                             values = 'customer_unique_id')
cohorts_pivot
```

```python
# рассчитаем Retention Rate, то есть долю клиентов, сделавших повторный заказ.
# для этого по каждой когорте разделим число клиентов, сделавших повторный заказ на размер соотвествующей когорты 
# из таблицы 'cohorts_size'. и выразим значения в процентах с огруглением до двух знаков после запятой.
retention_rate = cohorts_pivot.divide(cohorts_size, axis=0).mul(100).round(2)
# перименуем 'first_order' в 'Когорты'
retention_rate.index.names = ['Когорты']
retention_rate
```

```python
# для наглядности построим диаграмму когортного анализа
plt.rcParams['font.size'] = '12'
plt.figure(figsize=(12,9))
plt.title('Когортный анализ\nRetention Rate, %\n')

ax = sns.heatmap(data=retention_rate, annot=True, vmin=0.0,vmax=0.8 ,cmap='crest', linewidth=.5)
# добавим подпись оси абцисс
ax.set(xlabel="\nПорядковый номер периода анализа")

plt.show()
```

```python
# выведем данные когортного анализа на третий месяц наблюдений
print('Данные когортного анализа на третий месяц наблюдений:\n', retention_rate[3])

# как видно из анализа, максимальный retention имеет когорта клиентов, сделавших первый заказ '2017-09'
print('\nТаким образом, максимальный Retention Rate имеют пользователи из когорты', retention_rate[3].idxmax())
```

```python
# Задание 6.
# Часто для качественного анализа аудитории использую подходы, основанные на сегментации. 
# Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию.

# В кластеризации можешь выбрать следующие метрики: 
# R - время от последней покупки пользователя до текущей даты, 
# F - суммарное количество покупок у пользователя за всё время, 
# M - сумма покупок за всё время. 

# Подробно опиши, как ты создавал кластеры. Для каждого RFM-сегмента построй границы метрик recency, 
# frequency и monetary для интерпретации этих кластеров. 

# Пример такого описания: 
# RFM-сегмент 132 (recency=1, frequency=3, monetary=2) имеет границы метрик 
# recency от 130 до 500 дней, 
# frequency от 2 до 5 заказов в неделю, 
# monetary от 1780 до 3560 рублей в неделю.
```

```python
# добавим к данным по заказам и товарам данные по пользователям
e = purchased_goods.merge(cust_df, on = 'customer_id', how = 'left')
e
```

```python
# определим период активности пользователей
print('Orders from {} to {}'.format(purchased_goods['order_purchase_timestamp'].min(),
                                    purchased_goods['order_purchase_timestamp'].max()))
```

```python
# мы анализируем период продаж в 2 года
period = 730
```

```python
# посчитаем сумму каждого заказа 
orders = e \
    .groupby(['order_id', 'order_purchase_timestamp', 'customer_unique_id']) \
    .agg({'price':'sum'}) \
    .sort_values('order_purchase_timestamp') \
    .reset_index()
orders
```

```python
# точку отсчета обозначим на 1 день больше даты посленего заказа
zero_point = purchased_goods['order_purchase_timestamp'].max() + timedelta(days=1)
zero_point
```

```python
# определяем сколько дней прошло с последнего заказа (recency)
orders['days_from_order'] = (zero_point - orders.order_purchase_timestamp)
```

```python
orders
```

```python
# посчитаем общую сумму покупок для дальнейшей проверки
total_price_sum = orders.price.sum()
total_price_sum
```

```python
# для каждого customer_id определим время с последнего заказа и количество заказов
rfm = orders\
    .groupby('customer_unique_id') \
    .agg({'days_from_order':'min',
                 'order_id':'count',
                 'price':'sum'}) \
    .sort_values('days_from_order')\
    .rename(columns = {'days_from_order':'recency',
                              'order_id':'frequency',
                                 'price':'monetary'}) \
    .reset_index()
rfm
```

```python
# для оценки интервалов для границ кластеров воспользуемся методом describe 
rfm.describe(include = 'all')
```

```python
#  определим границы для кластеров по каждой метрике
# Разбиваем на на ТРИ группы. 1 - Отлично, 2 - Хорошо, 3 - Плохо

# R - время от последней покупки пользователя до текущей даты

# 1 : '1-30 days', 
# 2 : '31-365 days', 
# 3 : 'more than 1 year', 


# F - суммарное количество покупок у пользователя за всё время

# 1 : 'more than 5 orders', 
# 2 : '2-5 orders', 
# 3 : '1 order' 


# M - сумма покупок за всё время

# 1 : 'more than 155 ', (75-й процентиль)
# 2 : '50 - 155', 
# 3 : 'less than 50'(примерно 25-й процентиль)
```

```python
# добавляем колонку с группами по R
rfm['recency_cat'] = pd.cut(rfm.recency, \
                            bins=pd.to_timedelta(['0d', '30d', '365d', '730d']), \
                            labels = ['1', '2', '3'])
```

```python
# добавляем колонку с группами по F
rfm['frequency_cat'] = pd.cut(rfm.frequency, \
                            bins = [0, 1, 5, 100], \
                            labels = ['3', '2', '1'])
```

```python
# добавляем колонку с группами по M
rfm['monetary_cat'] = pd.cut(rfm.monetary, \
                            bins = [0, 50, 155, 100000], \
                            labels = ['3', '2', '1'])
```

```python
# Смотрим, что получилось
rfm.head()
```

```python
# изменим тип данных новых колонок для последующих операций объединения в кластер
rfm[['recency_cat', 'frequency_cat', 'monetary_cat']] = rfm[['recency_cat', 'frequency_cat', 'monetary_cat']].astype('int32')
```

```python
# проверяем что тип данных колонок с группами изменился на 'int32'
rfm.dtypes
```

```python
# сведем группы в колонку общей оценки
rfm['RFM_Score'] = rfm['recency_cat'].map(str) + rfm['frequency_cat'].map(str) + rfm['monetary_cat'].map(str)
# отсортируем оценки по возрастанию
rfm = rfm.sort_values('RFM_Score')
```

```python
rfm.head()
```

```python
# посчитаем доли RFM-сегментов 
rfm_top = rfm.RFM_Score.value_counts(normalize=True).mul(100).round(0).head()
rfm_top
```

```python
# 5 сегментов составляют более 80% от всей совокупности клиентов
# изобразим доли этих сегментов на круговой диаграмме
colors = sns.color_palette('pastel')[0:5]

plt.pie(rfm_top, colors = colors, autopct='%.0f%%', labels = rfm_top.index)
plt.title('Соотношение ТОП-5\nRFM-сегментов клиентов\n')
plt.show()
```

```python
# построим диаграмму количества клиентов по RFM-сегментам
plt.figure(figsize=(18,9))
sns.countplot(rfm.RFM_Score)
```
